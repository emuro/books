```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

# C2. Describing the Data

```python

head_tail = ["".join(random.choice("HT")) for i in range(0,1000)]
print(head_tail)
df = pd.DataFrame(head_tail, columns=["toss_coin"])
#print(df)
df.info()
```

```python
# setting font sizeto 30
plt.rcParams.update({'font.size': 30})

fig, ax = plt.subplots(figsize=(10,10))
ax = df.toss_coin.value_counts().plot.bar(color=["red", "blue"])
ax.set_title("Tossing a coin")
ax.set_ylabel("Number")
ax.set_xlabel("Event")
plt.show()
```
